```python
import numpy as np
from random import *
import math
import matplotlib.pyplot as plt
import networkx as nx 
```

# Generate adjacency matrix

```python
size = 100
ed = 200
array = np.zeros((size,size), dtype=int)

edges = 0

while edges<ed:
    edges+=1
    a = 1
    b = 1
    while a==b or array[a-1][b-1]==1:
        a = randint(1,size)
        b = randint(1,size)
    array[a-1][b-1] = 1
    array[b-1][a-1] = 1
```

```python
import pandas as pd
table = pd.DataFrame(data=array,columns=np.arange(1,array.shape[0]+1),index=np.arange(1,array.shape[1]+1))
# print(table.head())
table
```

# Generate adjacency list

```python
graph = {}
for i in range(1,size+1):
    s = []
    for j in range(1,size+1):
        if array[i-1][j-1] == 1:
            s.append(str(j))
    graph[str(i)] = s

for key, val in graph.items():
          print('{}:{}'.format(key, val))
```

# Visualizing graph 

```python
G = nx.Graph(array)

options = {
    "font_size": 0,
    "node_size": 10,
    "node_color": "blue",
    "edgecolors": "blue",
    "linewidths": 1,
    "width": 1,
}

nx.draw(G, pos=nx.spring_layout(G), **options)
plt.savefig(fname = "D:/Магистр/Algorithms/graph_5.1.png", dpi=300, facecolor='w', edgecolor='w',
            orientation='portrait', papertype=None, format=None, 
            transparent=False, bbox_inches=None, pad_inches=0.1,
            frameon=None, metadata=None)
```

```python
mapping = {}
for i in range(0,array.shape[0]):
    mapping[i] = i+1

G = nx.relabel_nodes(G, mapping)
```

```python
options = {
    "font_size": 6,
    "node_size": 100,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 1,
}

nx.draw_networkx(G, **options)

# Set margins for the axes so that nodes aren't clipped
#ax = plt.gca()
#ax.margins(0.2)
plt.savefig(fname = "D:/Магистр/Algorithms/graph_5.2.png", dpi=300, facecolor='w', edgecolor='w',
            orientation='portrait', papertype=None, format=None, 
            transparent=False, bbox_inches=None, pad_inches=0.1,
            frameon=None, metadata=None)
# plt.axis("off")
# plt.show()
```

```python
pos = nx.circular_layout(G)
# Create a figure with 1:1 aspect ratio to preserve the circle.
fig, ax = plt.subplots(figsize=(8, 8))
node_opts = {"node_size": 200, "node_color": "w", "edgecolors": "k", "linewidths": 1.0}
nx.draw_networkx_nodes(G, pos, **node_opts)
nx.draw_networkx_labels(G, pos, font_size=10)
# Extract color from edge data

nx.draw_networkx_edges(G, pos, width=1.0)

ax.set_axis_off()
fig.tight_layout()
plt.savefig(fname = "D:/Магистр/Algorithms/graph_5.3.png", dpi=300, facecolor='w', edgecolor='w',
            orientation='portrait', papertype=None, format=None, 
            transparent=False, bbox_inches=None, pad_inches=0.1,
            frameon=None, metadata=None)
```

# Depth-first search to find connected components of the graph

```python
S = [G.subgraph(c).copy() for c in nx.connected_components(G)]
for i in range(nx.number_connected_components(G)):
    print(S[i].nodes)
    print()
```

# Breadthfirst search to find a shortest path between two random vertices

```python
s = randint(1,size)
t = randint(1,size)
print(f'{s} --> {t}')
p = nx.shortest_path(G, s, t)
print(p)
```
