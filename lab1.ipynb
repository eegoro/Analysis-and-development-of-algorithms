 # Implementations Algorithms bubble and quick sorting

```python
def bubble_sort(nums):  
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                swapped = True    
def partition(nums, low, high):  
    pivot = nums[(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while nums[i] < pivot:
            i += 1
        j -= 1
        while nums[j] > pivot:
            j -= 1
        if i >= j:
            return j
        nums[i], nums[j] = nums[j], nums[i]
def quick_sort(nums):  
    def _quick_sort(items, low, high):
        if low < high:
            split_index = partition(items, low, high)
            _quick_sort(items, low, split_index)
            _quick_sort(items, split_index + 1, high)
    _quick_sort(nums, 0, len(nums) - 1)
```

# Time complexity algorithms with random vector

```python
max_n = 2000
time_n_iter_const = np.arange(max_n*5, dtype=float)
time_n_const = np.arange(max_n, dtype=float)
time_n_iter_const = time_n_iter_const.reshape((5, max_n))

time_n_sum = np.arange(max_n, dtype=float)
time_n_iter_sum = np.arange(max_n*5, dtype=float)
time_n_iter_sum = time_n_iter_sum.reshape((5, max_n))

time_n_prod = np.arange(max_n, dtype=float)
time_n_iter_prod = np.arange(max_n*5, dtype=float)
time_n_iter_prod = time_n_iter_prod.reshape((5, max_n))

max_n_poly = 1700
time_n_poly = np.arange(max_n_poly, dtype=float)
time_n_iter_poly = np.arange(max_n_poly*5, dtype=float)
time_n_iter_poly = time_n_iter_poly.reshape((5, max_n_poly))

max_n_bubble = 1000
time_n_bubble = np.arange(max_n_bubble, dtype=float)
time_n_iter_bubble = np.arange(max_n_bubble*5, dtype=float)
time_n_iter_bubble = time_n_iter_bubble.reshape((5, max_n_bubble))

time_n_quick = np.arange(max_n, dtype=float)
time_n_iter_quick = np.arange(max_n*5, dtype=float)
time_n_iter_quick = time_n_iter_quick.reshape((5, max_n))

time_n_timesort = np.arange(max_n, dtype=float)
time_n_iter_timesort = np.arange(max_n*5, dtype=float)
time_n_iter_timesort = time_n_iter_timesort.reshape((5, max_n))

fv = 0
for iter in range(5):
    for n in range(max_n):
        v = np.arange(n)
        v = np.random.randint(0, 100, n)
       
        time_n_iter_const[iter][n] = constfun(fv, v)
        time_n_iter_sum[iter][n] = sumfun(fv, v)
        time_n_iter_prod[iter][n] = prodfun(fv, v)
        if (n < max_n_poly):
            time_n_iter_poly[iter][n] = polyfun(fv, v)
        if (n < max_n_bubble):
            time_n_iter_bubble[iter][n] = bubblefun(fv, v)
        time_n_iter_quick[iter][n] = quickfun(fv, v)
        time_n_iter_timesort[iter][n] = timesortfun(fv, v)

time_n_const = middletime(max_n, time_n_iter_const, time_n_const)
time_n_sum = middletime(max_n, time_n_iter_sum, time_n_sum)
time_n_prod = middletime(max_n, time_n_iter_prod, time_n_prod)
time_n_poly = middletime(max_n_poly, time_n_iter_poly, time_n_poly)
time_n_bubble = middletime(max_n_bubble, time_n_iter_bubble, time_n_bubble)
time_n_quick = middletime(max_n, time_n_iter_quick, time_n_quick)
time_n_timesort = middletime(max_n, time_n_iter_timesort, time_n_timesort)

x_poly = np.arange(max_n_poly)
x_bubble = np.arange(max_n_bubble)
x_max = np.arange(max_n)
y_const = time_n_const
y_sum = time_n_sum
y_prod = time_n_prod
y_poly = time_n_poly
y_bubble = time_n_bubble
y_quick = time_n_quick
y_timesort = time_n_timesort

popt, _ = curve_fit(const, x_max, y_const)
a_const = popt
y_new_const = np.full(max_n, a_const)
pointgraph(x_max, x_max, y_const, y_new_const, 'Fitted f(x) = a', "D:/Магистр/Algorithms/const.png")

popt, _ = curve_fit(lin, x_max, y_sum)
a_sum, b_sum = popt
y_new_sum = lin(x_max,a_sum, b_sum)
pointgraph(x_max, x_max, y_sum, y_new_sum, 'Fitted f(x) = a$\cdot$x+b', "D:/Магистр/Algorithms/sum.png")

popt, _ = curve_fit(lin, x_max, y_prod)
a_prod, b_prod = popt
y_new_prod = lin(x_max,a_prod, b_prod)
pointgraph(x_max, x_max, y_prod, y_new_prod, 'Fitted f(x) = a$\cdot$x+b', "D:/Магистр/Algorithms/prod.png")

popt, _ = curve_fit(lin, x_poly, y_poly)
a_poly, b_poly = popt
y_new_poly = lin(x_max,a_poly, b_poly)
pointgraph(x_poly, x_max, y_poly, y_new_poly, 'Fitted f(x) = a$\cdot$x+b', "D:/Магистр/Algorithms/poly.png")

popt, _ = curve_fit(xpower2, x_bubble, y_bubble)
a_bubble = popt
y_new_bubble = xpower2(x_max,a_bubble)
pointgraph(x_bubble, x_max, y_bubble, y_new_bubble, 'Fitted f(x) = a$\cdot$x$^2$', "D:/Магистр/Algorithms/bubble.png")

popt, _ = curve_fit(xlogx, x_max, y_quick)
a_quick, b_quick = popt
y_new_quick = xlogx(x_max,a_quick, b_quick)
pointgraph(x_max, x_max, y_quick, y_new_quick, 'Fitted f(x) = a$\cdot$x$\cdot$log(x+b)', "D:/Магистр/Algorithms/quick.png")

popt, _ = curve_fit(xlogx, x_max, y_timesort)
a_timesort, b_timesort = popt
y_new_timesort = xlogx(x_max,a_timesort, b_timesort)
pointgraph(x_max, x_max, y_timesort, y_new_timesort, 'Fitted f(x) = a$\cdot$x$\cdot$log(x+b)', "D:/Магистр/Algorithms/timesort.png")
```

# Time complexity algorithms with random matrixs

```python

max_n = 300
time_n = np.arange(max_n, dtype=float)
time_n_iter = np.arange(max_n*5, dtype=float)
time_n_iter = time_n_iter.reshape((5, max_n))
for iter in range(5):
    for n in range(1, max_n):
        
        A = np.random.randint(0, 100, (n, n))

        B = np.random.randint(0, 100, (n, n))

        C = np.random.randint(0, 100, (n, n))
    
        start_time = time.time()
        for j in range(n):
            for i in range(n):
                sum = 0
                for k in range(n):
                    sum = sum + A[j][k]*B[k][i]
                C[j][i] = sum
                sum = 0
        time_n_iter[iter][n] = time.time() - start_time

for n in range(max_n):
    time_n[n] = (time_n_iter[0][n]+time_n_iter[1][n]+time_n_iter[2][n]+time_n_iter[3][n]+time_n_iter[4][n])/5
        
time_n[0] = 0

time_matrix = time_n

max_n_matrix = 300
x_matrix = np.arange(max_n_matrix)
x_max = np.arange(max_n)
y_matrix = time_matrix

popt, _ = curve_fit(xpower3, x_matrix, y_matrix)
a_matrix = popt
y_new_matrix = xpower3(x_max,a_matrix)
pointgraph(x_matrix, x_max, y_matrix, y_new_matrix, 'Fitted f(x) = a$\cdot$x$^3$', "D:/Магистр/Algorithms/matrix.png")
```
